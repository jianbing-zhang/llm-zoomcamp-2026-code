```python
from embedder import Embedder

embed = Embedder()

print(embed)
```

```python
q1 = 'How does approximate nearest neighbor search work?'
q1_vector = embed.encode(q1)
print(len(q1_vector), q1_vector[0])
```

```python
# load the class data
from gitsource import GithubRepositoryDataReader

reader = GithubRepositoryDataReader(
    repo_owner="DataTalksClub",
    repo_name="llm-zoomcamp",
    commit_id="8c1834d",
    allowed_extensions={"md"},
    filename_filter=lambda path: "/lessons/" in path,
)

documents = [file.parse() for file in reader.read()]
print(documents[0])
```

```python
#q2
q2_page = [document for document in documents if document.get('filename') == "02-vector-search/lessons/07-sqlitesearch-vector.md"]
q2_vector = embed.encode(q2_page[0].get('content'))
print(q1_vector.dot(q2_vector))
```

```python
#q3
from gitsource import chunk_documents
chunks = chunk_documents(documents, size=2000, step=1000)
texts = [chunk["filename"] + " " + chunk["content"] for chunk in chunks]

from tqdm.auto import tqdm
import numpy as np

batch_size = 50
X = []

for i in tqdm(range(0, len(texts), batch_size)):
    batch = texts[i:i + batch_size]
    batch_vectors = embed.encode_batch(batch)
    X.extend(batch_vectors)

X = np.array(X)
```

```python
scores = X.dot(q1_vector)
idx = np.argmax(scores)

chunks[idx]
```

```python
from minsearch import VectorSearch
vindex = VectorSearch(keyword_fields=["filename"])
vindex.fit(X, chunks)
```

```python
q4 = 'What metric do we use to evaluate a search engine?'
q4_vector = embed.encode(q4)
results = vindex.search(q4_vector, num_results=5)
```

```python
print(results[0])
```

```python
from minsearch import Index

text_index = Index(
    text_fields=["content"],
    keyword_fields=["filename"]
)

text_index.fit(chunks)
```

```python
# q5
q5 = "How do I store vectors in PostgreSQL?"

text_results = text_index.search(
    q5,
    boost_dict={"filename": 2.0},
    num_results=5
)
text_filenames = [result['filename'] for result in text_results]

q5_vector = embed.encode(q5)
v_results = vindex.search(q5_vector, num_results=5)
vs_filenames = [result['filename'] for result in vs_results]
print(set(vs_filenames), set(text_filenames))
```

```python
# q6
def rrf(result_lists, k=60, num_results=5):
    scores = {}
    docs = {}

    for results in result_lists:
        for rank, doc in enumerate(results):
            key = (doc["filename"], doc["start"])
            scores[key] = scores.get(key, 0) + 1 / (k + rank)
            docs[key] = doc

    ranked = sorted(scores, key=scores.get, reverse=True)
    return [docs[key] for key in ranked[:num_results]]
```

```python
q6 = "How do I give the model access to tools?"

text_q6_results = text_index.search(
    q6,
    boost_dict={"filename": 2.0},
    num_results=5
)

q6_vector = embed.encode(q6)
v_q6_results = vindex.search(q6_vector, num_results=5)

results = rrf([text_q6_results, v_q6_results])
print(results[0]["filename"])
```
